```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/data_allyields.csv')
data.head()
```

```python
data_A = data.drop(columns = ['Name', 'ID'])
corr_matrix_A = data_A.corr()
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
sns.heatmap(corr_matrix_A, annot=True, square=True, cmap='YlOrRd', annot_kws={'size': 18}, vmin=0, vmax=1)
ax.set_aspect('equal', adjustable='box')
plt.show()
fig.savefig("result/correlation_yield.pdf")
```
