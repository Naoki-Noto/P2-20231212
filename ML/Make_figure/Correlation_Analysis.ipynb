```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('data/correlation_iso.csv')
data.head()
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111)
sns.heatmap(data, annot=False, square=True, cmap="YlOrRd",vmin=0, vmax=1)
ax.set_aspect('equal', adjustable='box')
plt.show()
fig.savefig("result/correlation.pdf")
```
