```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/data_learning_curve.csv')
data.head()
```

```python
plt.figure(figsize=(6, 4))
for column in data.columns[1:]:
    plt.plot(data['training_data_count'], data[column], marker='o', label=column)

plt.xlim(8, 52)
plt.ylim(0.5, 0.8)
plt.xlabel('Training data count')
plt.ylabel('R2_score')
plt.legend()
plt.show()
```
