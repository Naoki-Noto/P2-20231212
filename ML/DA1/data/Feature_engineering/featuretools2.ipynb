```python
import pandas as pd
import featuretools as ft
```

```python
data = pd.read_csv('data_DFS1.csv')
data.head()
```

```python
es = ft.EntitySet(id="propertiy_data")
es = es.add_dataframe(dataframe_name='properties', dataframe=data, index='ID')
es['properties']
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_dataframe_name='properties',
                                      trans_primitives=['add_numeric', 'subtract_numeric', 'multiply_numeric', 'divide_numeric'],
                                      agg_primitives=[],
                                      max_depth=1)
feature_matrix
```

```python
feature_matrix = feature_matrix.drop(columns=['P(E_S1)', 'P(E_T1)', 'P(HOMO)', 'P(dDM)', 'P(dEST)'])
feature_matrix
```

```python
feature_matrix.to_csv('data_DFS2.csv', index=False)
```
