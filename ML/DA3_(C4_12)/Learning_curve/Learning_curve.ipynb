```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print('TrAB/DFT_FE model')
data1 = pd.read_csv('data/data_DFT_FE.csv')
print(data1)
print('===================================================================')
plt.figure(figsize=(5, 4))
plt.errorbar(data1['training_data_count'],
             data1['mean'],
             yerr=data1['std'],
             fmt='-o',
             capsize=3,
             c='b')

plt.xlim(7, 53)
plt.ylim(0.4, 0.9)
plt.xlabel('Training data count')
plt.ylabel('Avg R2')
#plt.savefig('TrAB_DFT_FE_model.pdf')
plt.show()
```

```python
print('TrAB/DFT_FE2 model')
data2 = pd.read_csv('data/data_DFT_FE2.csv')
print(data2)
print('===================================================================')
plt.figure(figsize=(5, 4))
plt.errorbar(data2['training_data_count'],
             data2['mean'],
             yerr=data2['std'],
             fmt='-o',
             capsize=3,
             c='r')

plt.xlim(7, 53)
plt.ylim(0.4, 0.9)
plt.xlabel('Training data count')
plt.ylabel('Avg R2')
#plt.savefig('TrAB_DFT_FE2_model.pdf')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(data1['training_data_count'], data1['mean'], marker='o', c='b', label='TrAB/DFT_FE')
plt.plot(data2['training_data_count'], data2['mean'], marker='o', c='r', label='TrAB/DFT_FE2')
plt.xlim(8, 52)
plt.ylim(0.5, 0.8)
plt.xlabel('Training data count')
plt.ylabel('Avg R2')
plt.legend(loc="upper left")
plt.show()
```
