```python
import pandas as pd
import numpy as np
import featuretools as ft
```

```python
data = pd.read_csv('data_OPS.csv')
data = data.drop(columns=['Name'])
ID = data.drop(columns=['HOMO', 'LUMO', 'f_S1', 'E_T1', 'dDM'])
data.head()
```

```python
es = ft.EntitySet(id="propertiy_data")
es = es.add_dataframe(dataframe_name='properties', dataframe=data, index='ID')
es['properties']
```

```python
feature_matrix, feature_defs = ft.dfs(entityset=es,
                                      target_dataframe_name='properties',
                                      trans_primitives=['percentile'],
                                      agg_primitives=[],
                                      max_depth=1)
feature_matrix
```

```python
feature_matrix = feature_matrix.drop(columns=['HOMO', 'LUMO', 'f_S1', 'E_T1', 'dDM'])
feature_matrix
```

```python
feature_matrix = np.array(feature_matrix)
feature_matrix = pd.DataFrame(feature_matrix, columns=['P(E_T1)', 'P(HOMO)', 'P(LUMO)', 'P(dDM)', 'P(f_S1)'])
feature_matrix = pd.concat([ID, feature_matrix], axis=1, join='inner')
feature_matrix
```

```python
feature_matrix.to_csv('data_DFS1.csv', index=False)
```
